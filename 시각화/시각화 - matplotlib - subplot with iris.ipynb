#### 1. 꽃잎, 꽃받침 수치를 데이터로, 속성명을 컬럼으로 가지는 데이터프레임 생성

```python
df = pd.DataFrame(dataset.data, columns=dataset['feature_names'])
df.head()
```

#### target 데이터와 target_names 데이터를 결합하여 종이름 데이터를 컬럼으로 추가

```python
# 각 수치형 target을 문자형/명목형 데이터로 변형하여 컬럼으로 결합
# 0 -> setosa, 1-> versicolor, 2-> virginica
# [(k, v), (k, v)] -> {k:v, k:v}
# zip([1, 2, 3], ['a', 'b', 'c']) -> {1:'a', 2:'b', 3:'c'}
species_map = zip([0, 1, 2], dataset['target_names'])
species_map = dict(species_map)
species_map
```

```python
dataset.target_names
```

```python
# 새로운 컬럼인 species 생성
# 컬럼 인덱싱 = 리스트
# 0 -> setosa로 표기, 1-> versicolor로 표기, 2->virginica로 표기해서
# df데이터의 ['species'] 컬럼으로 저장
# [species_map[i] for i in dataset['target']]
df['species'] = [species_map[i] for i in dataset['target']]
df.head()
```

### Pie Plot(원형 차트)
- 전체에서 특정 범주가 차지하는 비중/비율을 파악하기 위해 사용
- pandas.DataFrame/Series.plot(kind='pie')
    - Series 객체 : 하나의 원형 차트 생성
    - DataFrame 객체 : 각 열 마다 하나씩 원형 차트 생성
- pandas.DataFrame/Series.plot.pie(y=None, figsize=None, subplots=False)
    - y : 하나의 컬럼에 대해 원형 차트를 생성할 때 사용할 컬럼명
    - figsize : 생성할 figure 객체 크기(tuple)
    - subplots : y 값으로 하나의 컬럼을 사용하는 것이 아니라 전체 df를 사용할 때 True로 지정

```python
# 전체 종별 데이터 평균을 df 를 이용해 구해주세요.
# groupby로 하시고 변수명은 mean_by_species입니다. 
mean_by_species = df.groupby('species').mean()
```

```python
mean_by_species
```

```python
# 하나의 속성(feature) 별로 종별 파이차트 생성
# 속성별로 종별 비중
# subplots : 컬럼별로 subplot 생성 여부
# layout : 튜플 타입으로 행, 열 구조 지정
# legend : 범례 여부
mean_by_species.plot(kind='pie', subplots=True, layout=(2,2), legend=False)
plt.show()
```

```python
# 4개 컬럼 모두가 아닌 하나의 요소에 대해서만 파이플롯 생성시
# label : 사용한 컬럼명 표기 여부
# title : axes 자체의 title 지정
# y에 내가 그리고싶은 컬럼 하나만 지정
mean_by_species.plot(kind='pie', y='sepal length (cm)', legend=False, label='', title='<SEPAL LENGTH>')
plt.show()
```

### Bar plot(막대형 차트)

- pandas.DataFrame/Series.plot(kind='bar')
    - Series 객체 : 하나의 막대 차트 생성
    - DataFrame 객체 : 각 열 마다 하나씩 막대 차트 생성
- pandas.DataFrame/Series.plot.bar(x=None, y=None, rot=None, figsize=None)
    - x : x축을 기준으로 사용되는 데이터
    - y : y축을 기준으로 사용되는 데이터(막대로 표현되는 수치)
    - rot : x축 라벨 회전 각도(기본값=옆으로 누은 형태), 0/1/-1일 때 수직

```python
# 1번 ~ 5번 데이터에 대한 feature 값 비교
# 첫 번째 방법
df[:5].plot(kind='bar', legend=False, rot=0, figsize=(10, 5))
plt.show()
```

```python
# 두 번째 방법
df[:5].plot.bar(rot=0, figsize=(5, 5), grid=True)
plt.ylim(0, 7)
plt.show()
```

```python
# 하나의 컬럼에 대해서 차트 만들기(y에 값 줘서 만들기)
# y = 'petal length (cm)'
df[:5].plot(kind='bar', y='petal length (cm)', rot=0, legend=False)
plt.show()
```

### Histogram(히스토그램)

- pandas.DataFrame/Series.plot(kind='hist')

- pandas.DataFrame/Series.hist(bins=10)
    - bins : 히스토그램 구간 수를 나타내는 값으로 구간 수가 너무 작으면(=간격이 너무 넓으면) 세부 분포 파악이 어렵고, 구간 수가 너무 많으면 전체적인 분포 파악이 어렵다.

### box-plot(상자그림)

- pandas.DataFrame/Series.plot(kind='box')
- pandas.DataFrame/Series.boxplot()

```python
df.head()
```

```python
# 종별 꽃받침 길이 속성에 대한 box-plot
# 최종 구조 : 상위 컬럼(길이 속성), 하위 컬럼(종) / 인덱스(종개수)
```

```python
def get_index(num):
    return num % 50
```

```python
# idx 컬럼 생성
# 종별 0~49     예)98 -> 48, 148->48, 48->48
# apply 함수는 내부에 들어온 조건식을 이용해 컬럼 하나의 모든 값을 교환해준다.
df['idx'] = pd.Series(df.index).apply(get_index)
df.tail()
```

```python
# groupby를 이용해서 idx별로 묶어서 종별 요소 평균값을 비교해주세요.
df.groupby('idx').mean()
```

```python
# species로 묶어서 각 종별 petal length (cm)의 기술통계값을 확인해주세요.
# boxplot을 그려주시면 됩니다.

df.boxplot(column='petal length (cm)', by='species')

plt.show()
```
